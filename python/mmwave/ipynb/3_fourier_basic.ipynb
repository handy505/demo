```python
import numpy as np
from numpy import sin, pi

xs = np.linspace(0, 2*pi, 128)
y = 1*sin(xs*1 + 0) + 1/2*sin(xs*3 + 0) + 1/3*sin(xs*7+pi/4) + 1/2*sin(xs*29+pi/2)
    

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

plt.plot(xs, y)


yf = abs(np.fft.fft(y))
yf1 = yf/len(xs)
yf2 = yf1[range(int(len(xs)/2))]

xf = np.arange(len(y))
xf1 = xf
xf2 = xf[range(int(len(xs)/2))]

%matplotlib inline
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')


plt.subplot(311)
plt.plot(xs, y)
plt.title('Original wave')
 
plt.subplot(312)
plt.plot(xf2,yf2,'b')
plt.title('FFT)',fontsize=10,color='b')
 
plt.subplot(313)    
plt.bar(xf2,yf2)
plt.title('FFT)',fontsize=10,color='b')

plt.show()
```
