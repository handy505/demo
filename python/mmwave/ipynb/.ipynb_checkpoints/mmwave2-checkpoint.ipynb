```python
import numpy as np

A = 1
times = np.linspace(0, 4, 1024)
txfreqs = np.linspace(0, 16, 1024)
tou = 0.5


rxfreqs = txfreqs - 1
for i, f in enumerate(rxfreqs):
    if f < 0:
        rxfreqs[i] = 0
        

txthedas = 2 * np.pi * txfreqs * times
rxthedas = 2 * np.pi * rxfreqs * times

txys = [A * np.sin(theda) for theda in txthedas]
rxys = [A * np.sin(theda) for theda in rxthedas]

ifsignals = A * np.sin((2*np.pi*txfreqs - 2*np.pi*rxfreqs)*times + (txthedas - rxthedas))


%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
fig = plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

plt.subplot(311)
plt.plot(times[:512], txys[:512])
plt.plot(times[:512], rxys[:512])

plt.subplot(312)
plt.plot(times, txfreqs)
plt.plot(times, rxfreqs)

plt.subplot(313)
plt.plot(times, ifsignals)

```
