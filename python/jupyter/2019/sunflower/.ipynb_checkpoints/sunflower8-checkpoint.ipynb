### pandas載入csv時，遇到Parse error
pandas官方文件 https://pandas.pydata.org/pandas-docs/stable/api.html
因前兩個raw為header，是說明用，所以columns會與後面的資料沒有對齊導致錯誤，在pd.read_csv()時，加入header參數跳掉即可。



```python


import requests
import pandas as pd
import io
url = 'http://greenmet.cwb.gov.tw/data/map_csv/Display/Range/Forecast/2019011500/NOCWRF_03000_00_K01WEP_2019011500_0030.csv?0.8598516249420058'

r = requests.get(url)  
with open('raw/NOCWRF_03000_00_K01WEP_2019011500_0030.csv', 'wb') as f:
    f.write(r.content)
```

```python
#df = pd.read_csv('raw/NOCWRF_03000_00_K01WEP_2019011500_0030.csv', encoding='utf-8', header=3)
df = pd.read_csv('raw/NOCWRF_03000_00_K01WEP_2019011500_0030.csv', encoding='utf-8', skiprows=3)
df.shape
```
