# 微分的展示

微分是由極限的觀念而來。一點都不抽象，相當的具體！  
先弄懂斜率的意義  
再弄懂極限  
就可以接到微分  

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 原始函式
def f(x):
    return x**2

# 微分就是極限
def differential(func, x):
    h = 0.1 # 調整h的大小，就會對極限有感覺
    return (func(x+h) - func(x))/(x+h-x)

xs = range(0,10)
ys = [f(i) for i in xs]
dys = [differential(f, x) for x in xs]


plt.plot(ys)
plt.plot(dys)
plt.show()

# 後話
# 把微分理解為「變化率」、物理上就像「加速度」一樣。都有助於了解微分的意義。
# 但需注意數學本質上做為一種思想溝通的語言，並不需要具有物理意義。
# 如果剛好可以用物理來達，那也只是幫助我們了解這些數學背後所要表達的含意。
```
