硬幣隨機丟100次，計數正面出現次數，會靠近50次、重復做此實驗，結果將呈常態分怖、屬離散資料

```python
import matplotlib.pyplot as plt
import random
import collections
import statistics

def gen_random_list():    
    return [random.randint(0, 1) for _ in range(0, 100)]

b = [0 for _ in range(0, 100)]
c = [0 for _ in range(0, 100)]
d = [0 for _ in range(0, 100)]

for i in range(0, 1000):
    a = gen_random_list()
    b[a.count(1)] += 1
    
for i in range(0, 2000):
    a = gen_random_list()    
    c[a.count(1)] += 1

for i in range(0, 3000):
    a = gen_random_list()    
    d[a.count(1)] += 1

print('u={}, sd={}'.format(statistics.mean(b), round(statistics.pstdev(b), 2)))
print('u={}, sd={}'.format(statistics.mean(c), round(statistics.pstdev(c), 2)))

plt.plot(b)
plt.plot(c)
plt.plot(d)
plt.xlabel('Positive side times')
plt.ylabel('counts')
plt.show()
```
