```python
# 以時間為X軸製作亂數數列

import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
matplotlib.style.use('ggplot')
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
print(ts)
ts = ts.cumsum()
print(ts)
ts.plot.bar()

g = ts.plot.line()

ts.describe()
```
