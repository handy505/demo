# 相關系數

使用相關系統來檢測離群、需被注意的點  
用每個點跟點之間的距離(相關)，即可挑出需被關注的點  
但當維度愈來愈多後，會有「維度的詛咒」現像，距離變得愈來愈不明顯。

```python
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_csv('hpv1.csv')
df2 = pd.read_csv('hpv2.csv')
df3 = pd.read_csv('hpv3.csv')
df4 = pd.read_csv('hpv4.csv')
df5 = pd.read_csv('hpv5.csv')
df6 = pd.read_csv('hpv6.csv')
df7 = pd.read_csv('hpv7.csv')
df8 = pd.read_csv('hpv8.csv')

op1 = df1.iloc[:,1:3]
op2 = df2.iloc[:,1:3]
op3 = df3.iloc[:,1:3]
op4 = df4.iloc[:,1:3]
op5 = df5.iloc[:,1:3]
op6 = df6.iloc[:,1:3]
op7 = df7.iloc[:,1:3]
op8 = df8.iloc[:,1:3]

op1.head()
```

```python
op2.head()
```

```python
opx = pd.concat([op1, op2, op3, op4, op5, op6, op7, op8], axis=1)

plt.plot(op1)
plt.plot(op2)
plt.plot(op3)
plt.plot(op4)
plt.plot(op5)
plt.plot(op6)
plt.plot(op7)
plt.plot(op8)

plt.show()
```
