```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
%matplotlib inline
register_matplotlib_converters()
```

```python
filename = 'solardb\AnXin-2020-0505.db'
with sqlite3.connect(filename) as con:
    df = pd.read_sql('''select * from inverter_minutely''', con=con)
    
df.LoggedDatetime = pd.to_datetime(df.LoggedDatetime) # datetime的部份，需先做過一次轉換
```

```python
df.columns
```

```python
# 依DeviceID將df取出
df1 = df.query('DeviceID == 1')
df2 = df.query('DeviceID == 2')
df3 = df.query('DeviceID == 3')

df4 = df.query('DeviceID == 4')
df5 = df.query('DeviceID == 5')
df18 = df.query('DeviceID == 18')
df19 = df.query('DeviceID == 19')
```

```python
fig, ax = plt.subplots(4, 4, sharex='col', sharey='row')
fig.set_figheight(10)
fig.set_figwidth(15)

ax[0,0].plot(df1.LoggedDatetime, df1.DC1Voltage)
ax[0,0].set_title('Inv-1 V')
ax[1,0].plot(df1.LoggedDatetime, df1.DC1Current)
ax[1,0].set_title('Inv-1 I')

ax[0,1].plot(df2.LoggedDatetime, df2.DC1Voltage)
ax[0,1].set_title('Inv-2 V')
ax[1,1].plot(df2.LoggedDatetime, df2.DC1Current)
ax[1,1].set_title('Inv-2 I')

ax[0,2].plot(df3.LoggedDatetime, df3.DC1Voltage)
ax[0,2].set_title('Inv-3 V')
ax[1,2].plot(df3.LoggedDatetime, df3.DC1Current)
ax[1,2].set_title('Inv-3 I')


ax[0,3].plot(df4.LoggedDatetime, df4.DC1Voltage)
ax[0,3].set_title('Inv-4 V')
ax[1,3].plot(df4.LoggedDatetime, df4.DC1Current)
ax[1,3].set_title('Inv-4 I')

ax[2,0].plot(df5.LoggedDatetime, df5.DC1Voltage)
ax[2,0].set_title('Inv-5 V')
ax[3,0].plot(df5.LoggedDatetime, df5.DC1Current)
ax[3,0].set_title('Inv-5 I')

ax[2,1].plot(df18.LoggedDatetime, df18.DC1Voltage)
ax[2,1].set_title('Inv-18 V')
ax[3,1].plot(df18.LoggedDatetime, df18.DC1Current)
ax[3,1].set_title('Inv-18 I')

ax[2,2].plot(df19.LoggedDatetime, df19.DC1Voltage)
ax[2,2].set_title('Inv-19 V')
ax[3,2].plot(df19.LoggedDatetime, df19.DC1Current)
ax[3,2].set_title('Inv-19 I')
```
