```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
%matplotlib inline
register_matplotlib_converters()
```

```python
filename = '../data/demo_2020_0505.sqlite'

with sqlite3.connect(filename) as con:
    invdf  = pd.read_sql('''select * from inverter_minutely''', con=con)
invdf.LoggedDatetime  = pd.to_datetime(invdf.LoggedDatetime)
```

```python
invdf = invdf.set_index('LoggedDatetime')
```

```python
invs = {}
for id in (1,2,3,4,5,18,19):
    invs[id] = invdf.query('DeviceID == {}'.format(id))

shapes = [inv.shape for inv in invs.values()]
print(shapes)
```

```python
def get_inv_mean(date_str, id):
    data = []
    #for k, t in interval.items():
    for h in range(5, 18):  
        start = '{} {:02d}:00:00'.format(date_str, h)
        end   = '{} {:02d}:00:00'.format(date_str, h+1)
        hourly = invs[id][start:end] # query rows by hour
        colname = 'mean-{}'.format(id)
        d = {'hour': h, colname: hourly.ACOutputPower.mean()}
        data.append(d)
    df = pd.DataFrame(data)
    df = df.set_index('hour')
    return df


def get_all_inv_mean(date_str):
    result = [get_inv_mean(date_str, id) for id in (1,2,3,4,5,18,19)]
    result = pd.concat(result, axis=1)
    return result

df = get_all_inv_mean('2020-05-05')
df
```
