```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
%matplotlib inline
register_matplotlib_converters()
```

```python
filename = '../data/demo_2020_0505.sqlite'
sql = '''SELECT DeviceID,
                strftime("%Y-%m-%d %H:%M:00", LoggedDatetime) AS minute,
                ACOutputPower
         FROM inverter_minutely'''
with sqlite3.connect(filename) as con:
    df = pd.read_sql(sql, con=con)
df.minute = pd.to_datetime(df.minute) # datetime的部份，需先做過一次轉換
df
```

```python
df.columns
```

```python
df.set_index('minute', inplace=True)
print(df.shape)
df.head()
```

```python
result = []
for id in (1,2,3,4,5,18,19):
    where_condition = df['DeviceID'] == id
    invdf = df[where_condition]
    invdf = invdf[['ACOutputPower']]
    invdf = invdf.rename(columns={'ACOutputPower':'power-{}'.format(id)})
    result.append(invdf)
df = pd.concat(result, axis=1)
print(df.to_string())
```

```python
df = df.resample('1H').mean()
df
```

```python
df = df.reset_index()
df['hour'] = df['minute'].dt.hour
df
df.set_index('hour', inplace=True)
df = df.drop(['minute'], axis=1)
df
```

```python
import matplotlib
matplotlib.style.use("ggplot")
fig, ax = plt.subplots(1)
df.plot()
```
