```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

iris = datasets.load_iris()
#print(type(iris))
#print(iris.keys())
#print(iris.data)
#print(iris.feature_names)

x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
#print(x.to_string())
y = pd.DataFrame(iris['target'], columns=['target'])
#print(y.to_string())
iris_data = pd.concat([x,y], axis=1)
#print(iris_data.to_string())
iris_data
```

```python
iris['target_names']
print(iris_data.columns)
```

```python
target_name = {
    0:'setosa',
    1:'versicolor',
    2:'virginica'
}
```

```python
iris_data['target_name'] = iris_data['target'].map(target_name)
iris_data = iris_data[(iris_data['target_name'] == 'setosa')|(iris_data['target_name'] == 'versicolor')]
iris_data = iris_data[['sepal length (cm)', 'petal length (cm)', 'target_name']]
print(iris_data.to_string())
```

```python
target_class = {
    'setosa':1,
    'versicolor':-1
}
```

```python
iris_data['target_class'] = iris_data['target_name'].map(target_class)
print(iris_data.to_string())
#iris.head(5)
```

```python
del iris_data['target_name']
print(iris_data.to_string())
```

```python
def sign(z):
    if z > 0:
        return 1
    else:
        return -1
```

```python
w = np.array([0.,0.,0.])
error = 1
iterator = 0
while error != 0:
    error = 0
    for i in range(len(iris_data)):
        x,y = np.concatenate((np.array([1.]), np.array(iris_data.iloc[i])[:2])), np.array(iris_data.iloc[i])[2]
        
        if sign(np.dot(w,x)) != y:
            print('-----------------')
            print("iterator: {}".format(iterator))
            print('DEBUG1 x: {}, w: {}, y: {}'.format(x, w, y))
            iterator += 1
            error += 1
            sns.lmplot('sepal length (cm)','petal length (cm)',data=iris_data, fit_reg=False, hue ='target_class')
            
            # Last Normal Vector
            if w[1] != 0:
                a = w[1]
                b = w[2]
                c = w[0]
                x_last_decision_boundary = np.array([0, a])
                y_last_decision_boundary = (b/a) * x_last_decision_boundary
                plt.plot(x_last_decision_boundary, y_last_decision_boundary,'c--')
            
            w = w + (y*x)
            
            print('DEBUG2 x: {}, w: {}, y: {}'.format(x, w, y))
            
            plt.plot([-10, 10], [0, 0], 'gray')
            plt.plot([0, 0], [-10, 10], 'gray')
            plt.plot(x[1], x[2], 'o', color='black')
            
            a = w[1]
            b = w[2]
            c = w[0]
            
            # Direction Vector
            x_decision_boundary = np.array([-5, 5])
            y_decision_boundary = (-a/b) * x_decision_boundary + (-c/b)
            plt.plot(x_decision_boundary, y_decision_boundary,'r')
            
            # Normal Vector
            x_decision_boundary_normal_vector = np.array([0, a])
            y_decision_boundary_normal_vector = (b/a) * x_decision_boundary_normal_vector
            plt.plot(x_decision_boundary_normal_vector, y_decision_boundary_normal_vector,'g')
            
            plt.xlim(-10,10)
            plt.ylim(-10,10)
            plt.show()
```
