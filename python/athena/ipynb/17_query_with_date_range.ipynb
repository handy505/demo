```python
import sqlite3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
%matplotlib inline
register_matplotlib_converters()
```

```python
filename = '..\solardb\AnXin-2020-0505.db'
with sqlite3.connect(filename) as con:
    irradf  = pd.read_sql('''select * from imeter_minutely''', con=con)
irradf.LoggedDatetime  = pd.to_datetime(irradf.LoggedDatetime)
irradf = irradf.set_index('LoggedDatetime')
```

```python
irradf.index
```

```python
datetimes1 = pd.date_range('2020-05-05 05:00:00', periods=12*13, freq='5T')
datetimes2 = pd.date_range('2020-05-05 05:05:00', periods=12*13, freq='5T')

datetimes2
```

```python
for dts in zip(datetimes1, datetimes2):
    
    df = irradf[dts[0]:dts[1]]
    print('{}, {}, {:05.1f}, {:05.3f}'.format(dts[0], dts[1], df.Value.mean(), df.Value.std()))
```

```python
fig, ax = plt.subplots(1, 2, sharex='col', sharey='row')
fig.set_figheight(5)
fig.set_figwidth(15)
```
